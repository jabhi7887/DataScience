```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Reading the dataset

```python
df=pd.read_csv('mpg_raw.csv')
```

# Bar plot

```python
sns.barplot(x='cylinders',y='mpg', data=df)
```

```python
df.columns
```

## Scatter plot

```python
plt.scatter(df['displacement'],df['horsepower'])
plt.xlabel('Displacement')
plt.ylabel('Horse power')
plt.title('Scatter plot ')
```

## Count plot

```python
sns.countplot(x='model_year',data=df)
plt.xlabel('Model year')
plt.title('Count plot ')
```

## Point plot

```python
sns.pointplot(x='cylinders', y='mpg', data=df)
```
